```python
import numpy as np
import matplotlib.pyplot as plt
import uproot, glob
```

```python
sim_files = glob.glob("scint_sim_t*.root")
```

```python
def parse_events(f):

    out_dict = {}

    energy_deps = np.array(f["T3"])
    decay_products = np.array(f["RDecayProducts"])

    event_ids = np.unique(energy_deps[-1,:])

    for evt in event_ids:
        mask = energy_deps[-1,:] == evt

        evt_energy_deps_e = energy_deps[0, mask]        
        evt_energy_deps_t = energy_deps[1, mask]
        evt_energy_deps_x = energy_deps[3, mask]
        evt_energy_deps_y = energy_deps[4, mask]
        evt_energy_deps_z = energy_deps[5, mask]
        energy_dep = np.vstack((evt_energy_deps_e, evt_energy_deps_t, evt_energy_deps_x, evt_energy_deps_y, evt_energy_deps_z)).T

        decay_mask = decay_products[-1,:] == evt
        curr_PID = decay_products[0, decay_mask]
        curr_Z = decay_products[1, decay_mask]
        curr_A = decay_products[2, decay_mask]
        unique_Z = np.unique(curr_Z)
        unique_A = np.unique(curr_A)

        print("Event", evt)
        track_ids = np.unique(energy_deps[-2, mask]) 
        print("Eng dep Track IDs", track_ids)
        track_ids = np.unique(decay_products[-2, decay_mask]) 
        print("Decay prod Track IDs", track_ids)
        print("Decay prod particle types: ")
        for tid in track_ids:
            tid_mask = decay_products[-2, decay_mask] == tid
            print(tid, decay_products[0, decay_mask][tid_mask], decay_products[4, decay_mask][tid_mask])
            eng_tid_mask = energy_deps[-2, mask] == tid
            print("energy deps")
            print(energy_dep[eng_tid_mask,:])


        if(len(unique_Z) > 1):
            print("Multiple Z values in event", evt)
            print(unique_Z)
        if(len(unique_A) > 1):
            print("Multiple A values in event", evt)
            print(unique_A)

        return None
        curr_dict = {"eng_dep": energy_dep, "Z": unique_Z[0], "A": unique_A[0]}

        out_dict[evt] = curr_dict


    return out_dict
```

```python
event_dict = {}
for sf in sim_files:

    curr_dict = parse_events(uproot.open(sf))
```
